```python
#Part 1

import numpy as np

# Data points 
X = np.array([1, 2, 3, 4, 5])
Y = np.array([1.2, 2.8, 3.6, 4.5, 5.1])
# Stack the design matrix with a column of ones (for the intercept)
A = np.vstack([X, np.ones(len(X))]).T
# Solve using least squares
m, c = np.linalg.lstsq(A, Y, rcond=None)[0]
print(f"Slope: {m}, Intercept: {c}")

import matplotlib.pyplot as plt

plt.scatter(X,Y)
plt.plot(np.linspace(0,5,50), m*np.linspace(0,5,50)+c)
plt.show()
```

```python
#Part 2

X1 = 10*np.random.rand(10)
X1.sort()
Y1 = 10*np.random.rand(10)
A1 = np.vstack([X1, np.ones(len(X1))]).T
print("X:\n",X1,"\nY:\n",Y1)
m1, c1 = np.linalg.lstsq(A1, Y1, rcond=None)[0]
print(f"\nSlope: {m1}, Intercept: {c1}")
plt.scatter(X1,Y1)
plt.plot(np.linspace(0,10,100), m1*np.linspace(0,10,100)+c1)
plt.show()
```

The line sort of fits the data, but because the data is random it is hard to approximate it very well with a line.

```python
def lineEq (X, m, c):
    return X*m + c

X = np.linspace(0,1,100)
Y = lineEq(X, 5, 3)
Y = Y + (np.random.random(100)*2 - 1)
A = np.vstack([X, np.ones(len(X))]).T
m, c = np.linalg.lstsq(A, Y, rcond=None)[0]
print(f"Approximate slope: {m}, Approximate intercept: {c}")
plt.scatter(X,Y)
plt.plot(X, 5*X+3, color='blue', label=f'true line')
plt.plot(X, m*X+c, color='red', label=f'least squares')
plt.legend()
plt.show()
residuals = abs(Y-(X*5+3))
print("residuals:: ", residuals, "\n largest residual", max(residuals), "\n smallest residual", min(residuals))
```
