```python
import numpy as np
import time

def jacobi(A, b, x0, tol, max_iterations):
    n = len(b)
    x = x0.copy()
    for k in range(max_iterations):
        x_new = np.zeros_like(x)
        for i in range(n):
            s = sum(A[i][j] * x[j] for j in range(n) if j != i)
            x_new[i] = (b[i] - s) / A[i][i]
        # Check for convergence
        if np.linalg.norm(x_new - x, ord=np.inf) < tol:
            return x_new, k
        x = x_new
    return x, max_iterations

def gauss_seidel(A, b, x0, tol, max_iterations):
    n = len(b)
    x = x0.copy()
    for k in range(max_iterations):
        x_new = x.copy()
        for i in range(n):
            s1 = sum(A[i][j] * x_new[j] for j in range(i)) # Using already updated values
            s2 = sum(A[i][j] * x[j] for j in range(i + 1, n)) # Using old values
            x_new[i] = (b[i] - s1 - s2) / A[i][i]
        # Check for convergence
        if np.linalg.norm(x_new - x, ord=np.inf) < tol:
            return x_new, k
        x = x_new
    return x, max_iterations

A = np.array([
[0.582745, 0.48, 0.10, 0., 0.], 
 [0.48, 1.044129, 0.46, 0.10, 0.],
 [0.10, 0.46, 1.10431 , 0.44, 0.10],
 [0., 0.10, 0.44, 0.963889, 0.42],
 [0., 0., 0.10	, 0.42	, 0.522565]], dtype=float)
b = np.array([1.162745, 2.084129, 2.20431 , 1.923889, 1.042565], dtype=float)
x0 = np.zeros_like(b)
tol = 1e-6
max_iterations = 100
timeJ = time.time()
xJ, iterationsJ = jacobi(A,b,x0,tol,max_iterations)
timeJ = time.time() - timeJ
timeGS = time.time()
xGS, iterationsGS = gauss_seidel(A,b,x0,tol,max_iterations)
timeGS = time.time() - timeGS
print(xJ, iterationsJ, timeJ)
print(xGS, iterationsGS, timeGS)
```

```python
def diagonallyDominant(A):
    n = A.shape[0]
    for i in range (n):
        if(abs(A[i,i]) < sum(np.abs(A[i]))-abs((A[i,i]))):
            return False
    return True

diagDom = diagonallyDominant(A)
print(A)
print('\nIs matrix A diagonally dominant:: ', diagDom)
```
