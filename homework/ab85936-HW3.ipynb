```python
import math
import numpy as np
import matplotlib.pyplot as plt

def taylorSin(numTerms,x):
    t = 0
    for i in range(0,numTerms):
        t += (((-1)**i)*(x**(2*i+1))) / (math.factorial(2*i+1))
    return t

oneTerm = np.empty([101])
twoTerm = np.empty([101])
threeTerm = np.empty([101])
fourTerm = np.empty([101])
i = np.linspace(0, 10, 101)
for x in range (0, 101):
    oneTerm[x] = abs(np.sin(x/10)-taylorSin(1, x/10))
    twoTerm[x] = abs(np.sin(x/10)-taylorSin(2, x/10))
    threeTerm[x] = abs(np.sin(x/10)-taylorSin(3, x/10))
    fourTerm[x] = abs(np.sin(x/10)-taylorSin(4, x/10))
plt.plot(i, oneTerm, 'r')
plt.plot(i, twoTerm, 'b')
plt.plot(i, threeTerm, 'k')
plt.plot(i, fourTerm, 'g')
plt.ylabel('Absolute Error')
plt.show()
```

The error seems to increase exponentially as x increases to larger and larger values. Additionally, it increases quickest for one term used in the approximation and a lot less quickly the more terms are used to approximate it.
