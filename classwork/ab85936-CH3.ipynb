```python
import random
import math
import numpy as np
import matplotlib.pyplot as plt

def approximatePi(x):
    counter = 0
    for i in range(0,x):
        a = random.uniform(-1,1)
        b = random.uniform(-1,1)
        if(a*a + b*b <= 1):
            counter += 1
    return(4*counter/x)

x = np.empty([2,100])
y = 0
for i in range (0,100):
    y = approximatePi(1000)
    x[0,i] = abs(math.pi-y)
    x[1,i] = abs((math.pi-y)/math.pi)
plt.subplot(1,2,1)
plt.scatter(range(100), x[0], s=10)
plt.title("Absolute Error")
plt.subplot(1,2,2)
plt.scatter(range(100), x[1], s=10)
plt.title("Relative Error")
```
