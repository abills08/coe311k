```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 2*np.pi, 50)
```

```python
import math 
import random

plt.subplot(2,2,1)
plt.plot(x, np.sin(x), 'r')
plt.subplot(2,2,2)
plt.scatter(x, np.cos(x))
plt.subplot(2,2,3)
z = np.empty(10)
for i in range (0,10):
    z[i] = random.uniform(-1,1)
plt.hist(z, 10)
plt.subplot(2,2,4)
plt.stem(x, np.sin(x), 'k')
```
