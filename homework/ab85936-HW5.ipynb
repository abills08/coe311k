```python
import numpy as np
import matplotlib.pyplot as plt

def myLeastSquared (X, Y):
    sumX = sum(X)
    sumY = sum(Y)
    sumX2 = sum(X*X)
    sumXY = sum(X*Y)
    m = (len(X)*sumXY-sumX*sumY)/(len(X)*sumX2-(sumX**2))
    c = (sumY-m*sumX)/len(X)
    return m,c

X = 10*np.random.rand(10)
X.sort()
Y = 10*np.random.rand(10)
print("X:\n",X,"\nY:\n",Y)
m, c = myLeastSquared(X, Y)
print(f"\nSlope: {m}, Intercept: {c}")
plt.scatter(X,Y)
plt.plot(np.linspace(0,10,100), m*np.linspace(0,10,100)+c)
plt.show()

X1 = X
Y1 = Y
A1 = np.vstack([X1, np.ones(len(X1))]).T
print("X:\n",X1,"\nY:\n",Y1)
m1, c1 = np.linalg.lstsq(A1, Y1, rcond=None)[0]
print(f"\nSlope: {m1}, Intercept: {c1}")
plt.scatter(X1,Y1)
plt.plot(np.linspace(0,10,100), m1*np.linspace(0,10,100)+c1)
plt.show()
```

```python
def lineEq (X, m, c):
    return X*m + c

X = np.linspace(0,1,100)
Y = lineEq(X, 5, 3)
Y = Y + (np.random.random(100)*2 - 1)
m, c = myLeastSquared(X,Y)
print(f"Approximate slope: {m}, Approximate intercept: {c}")
plt.scatter(X,Y)
plt.plot(X, 5*X+3, color='blue', label=f'true line')
plt.plot(X, m*X+c, color='red', label=f'least squares')
plt.legend()
plt.show()
residuals = abs(Y-(X*5+3))
print("residuals:: ", residuals, "\n largest residual", max(residuals), "\n smallest residual", min(residuals))
```
