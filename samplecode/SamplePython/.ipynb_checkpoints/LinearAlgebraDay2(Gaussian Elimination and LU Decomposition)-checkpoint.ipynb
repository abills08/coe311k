Goal of Gaussian Elimination: (same as np.linalg.solve())
- get leading coefficients for diagonal and 0s for all others below

Forward elimination:
1) id the pivot element
2) swap rows (if necessary)
3) normalize the pivot row
4) eliminate lower elements
5) move to the next column and repeat

Backward Substitution:
1) start with last equation
2) sub and solve
3) continue until all equations are solved

Interpret Results:
1) check for consistency
2) unique or infinite solutions

```python
import numpy as np

def gaussian_elimination(A, b):
    n = len(b)
    # Forward elimination
    for i in range(n):
        # Pivoting
        max_row = max(range(i, n), key=lambda x: abs(A[x][i]))
        A[i], A[max_row] = A[max_row], A[i]
        b[i], b[max_row] = b[max_row], b[i]
        # Make the rows below this one 0 in current column
        for j in range(i + 1, n):
            ratio = A[j][i] / A[i][i]
            A[j] = [A[j][k] - ratio * A[i][k] for k in range(n)]
            b[j] -= ratio * b[i]
    # Backward substitution
    x = [0] * n
    for i in range(n - 1, -1, -1):
        x[i] = (b[i] - sum(A[i][j] * x[j] for j in range(i + 1, n))) / A[i][i]
    return x

# Example Usage
A = [[3, 2, -4], [2, 3, 3], [5, -3, 1]]
b = [3, 15, 14]
solution = gaussian_elimination(A, b)
print(f"Solution: {solution}")
```

LU Decomposition:
Breaking a matrix A into two simpler matrices, A = LU, where L = a lower triangle matrix and U = upper triangle matrix

Lets you solve systems much faster by solving:
Lx = b
Ux = y

Decompose A into L and U:
- start with first element and create zeroes below diagonal in each column

LU Decomposition Algorithm
1) Initialize L as an identity matrix and U as a copy of A.
2) For each column:
    - Perform elimination below the pivot element to create zeros in the U matrix.
    - Store the multipliers used for elimination in the L matrix.

```python
import numpy as np

def lu_decomposition(A):
    n = len(A)
    L = np.eye(n) # Initialize L as an identity matrix
    U = np.array(A, dtype=float) # Copy of A to avoid
    # modifying the original matrix
    # modifying the original matrix
    for i in range(n):
        # Eliminate elements below the pivot (U[i, i])
        for j in range(i+1, n):
            if U[i, i] == 0:
                raise ValueError("Zero pivot encountered. LU decomposition cannot proceed.")
            factor = U[j, i] / U[i, i] # Multiplier for row
            # operation
            L[j, i] = factor # Store multiplier in L
            # Update the rows of U
            for k in range(i, n):
                U[j, k] -= factor * U[i, k]
    return L, U

# Example usage
A = np.array([[2, 3, 1], [4, 7, -1], [-2, 1, 2]])
L, U = lu_decomposition(A)
print("Matrix A:\n", A)
print("Lower Triangular Matrix L:\n", L)
print("Upper Triangular Matrix U:\n", U)
```

```python
import numpy as np
from scipy.linalg import lu

# Perform LU Decomposition
P, L, U = lu(A)
print("Matrix A:\n", A)
print("Permutation Matrix P:\n", P)
print("Lower Triangular Matrix L:\n", L)
print("Upper Triangular Matrix U:\n", U)
```
