```python
import numpy as np
import time

def jacobi(A, b, x0, tol, max_iterations):
    n = len(b)
    x = x0.copy()
    for k in range(max_iterations):
        x_new = np.zeros_like(x)
        for i in range(n):
            s = sum(A[i][j] * x[j] for j in range(n) if j != i)
            x_new[i] = (b[i] - s) / A[i][i]
        # Check for convergence
        if np.linalg.norm(x_new - x, ord=np.inf) < tol:
            return x_new, k
        x = x_new
    return x, max_iterations

def gauss_seidel(A, b, x0, tol, max_iterations):
    n = len(b)
    x = x0.copy()
    for k in range(max_iterations):
        x_new = x.copy()
        for i in range(n):
            s1 = sum(A[i][j] * x_new[j] for j in range(i)) # Using already updated values
            s2 = sum(A[i][j] * x[j] for j in range(i + 1, n)) # Using old values
            x_new[i] = (b[i] - s1 - s2) / A[i][i]
        # Check for convergence
        if np.linalg.norm(x_new - x, ord=np.inf) < tol:
            return x_new, k
        x = x_new
    return x, max_iterations

A = np.array([[10,-1,2,0,0],[-1,11,-1,3,0],[2,-1,10,-1,0],[0,3,-1,8,-2],[0,0,0,-2,9]])
b = np.array([14,30,26,25,37])
x0 = np.zeros_like(b)
tol = 1e-6
max_iterations = 100
timeJ = time.time()
xJ, iterationsJ = jacobi(A,b,x0,tol,max_iterations)
timeJ = time.time() - timeJ
timeGS = time.time()
xGS, iterationsGS = gauss_seidel(A,b,x0,tol,max_iterations)
timeGS = time.time() - timeGS
print(iterationsJ, timeJ)
print(iterationsGS, timeGS)
```

Both methods converged, Gauss-Seidel took less time but did more iterations

```python
A = np.array([[1,2,3,0,0],[2,1,2,3,0],[3,2,1,2,3],[0,3,2,1,2],[0,0,3,2,1]])
b = np.array([14,22,33,26,22])
x0 = np.zeros_like(b)
tol = 1e-6
max_iterations = 100
timeJ = time.time()
xJ, iterationsJ = jacobi(A,b,x0,tol,max_iterations)
timeJ = time.time() - timeJ
timeGS = time.time()
xGS, iterationsGS = gauss_seidel(A,b,x0,tol,max_iterations)
timeGS = time.time() - timeGS
print(iterationsJ, timeJ)
print(iterationsGS, timeGS)
```

Neither method converged because the array was not diagonally dominant.
