```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(0)
X = 10*np.random.rand(10)
X.sort()
Y = 10*np.random.rand(10)
A = np.vstack([X, np.ones(len(X))]).T
print("X:\n",X,"\nY:\n",Y)
m, c = np.linalg.lstsq(A, Y, rcond=None)[0]
print(f"\nSlope: {m}, Intercept: {c}")
mse = np.mean((Y - (X*m+c))**2)
print(f'Least Squared Mean Squared Error: {mse:.4f}')
ss_total = np.sum((Y - np.mean(Y))**2)
ss_res = np.sum((Y - (X*m+c))**2)
r_squared = 1 - (ss_res / ss_total)
print(f'Least Sqaured R-squared: {r_squared:.4f}\n')
degree = 3
coefficients = np.polyfit(X, Y, degree)
print("Polynomial Coefficients: ", coefficients)
polynomial = np.poly1d(coefficients)
y_pred = polynomial(X)
mse = np.mean((Y - y_pred)**2)
print(f'Polynomial Mean Squared Error: {mse:.4f}')
ss_total = np.sum((Y - np.mean(Y))**2)
ss_res = np.sum((Y - y_pred)**2)
r_squared = 1 - (ss_res / ss_total)
print(f'Polynomial R-squared: {r_squared:.4f}')
plt.scatter(X,Y)
plt.plot(X, y_pred, color='red', label=f'Polynomial fit of degree {degree}')
plt.plot(np.linspace(0,10,100), m*np.linspace(0,10,100)+c, color='blue', label=f'least squares')
plt.legend()
plt.show()
```

```python
degree = 3
V = np.vander(X, N=degree+1, increasing=True)
VT_V = np.dot(V.T, V)
VT_y = np.dot(V.T, Y)
coefficients = np.linalg.solve(VT_V, VT_y)
y_pred = coefficients[0]+coefficients[1]*X+coefficients[2]*X*X+coefficients[3]*X*X*X
mse = np.mean((Y - y_pred)**2)
print(f'Mean Squared Error: {mse:.4f}')
ss_total = np.sum((Y - np.mean(Y))**2)
ss_res = np.sum((Y - y_pred)**2)
r_squared = 1 - (ss_res / ss_total)
print(f'R-squared: {r_squared:.4f}\n')
plt.scatter(X,Y)
plt.plot(X, y_pred, color='red')
plt.show()
```
