```python
import random
import math
import numpy as np
import matplotlib.pyplot as plt

def approximatePi(x):
    counter = 0
    for i in range(0,x):
        a = random.uniform(-1,1)
        b = random.uniform(-1,1)
        if(a*a + b*b <= 1):
            counter += 1
    return(4*counter/x)

x = np.empty(500)
y = 0
z = np.empty(500)
j = 0
for i in range(10,10000,20):
    y = approximatePi(i)
    x[j] = abs(math.pi-y)
    z[j] = i
    j += 1
plt.scatter(z, x, s=10)
plt.title("Absolute Error")
plt.show()
#chatgpt used to check code for errors at end - changed x axis from 0-500 to 10-10000 as a result
```
