```python
import random

def approximatePi(x):
    counter = 0
    for i in range(0,x):
        a = random.uniform(-1,1)
        b = random.uniform(-1,1)
        if(a*a + b*b <= 1):
            counter += 1
    return(4*counter/x)
```

```python
approximatePi(5000)
```
